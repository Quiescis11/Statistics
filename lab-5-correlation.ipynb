# Imports and Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Current Power

```python
df = pd.read_csv(r'C:\Users\LENOVO\OneDrive\Documents\Github\Statistics\current-power.csv',
                delimiter=",")
df
```

```python
df.describe
```

```python
df.cov()
```

```python
#scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
# correlation coefficient
df.corr()
```

```python
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color = "#241c1c",
            alpha = 0.6,
            label='Data Points')

#line plot
plt.plot(x, line,
         color = "#128262",
         label=f'Regression Line (r = {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Air Quality

```python
air = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Documents\Github\Statistics\air-quality.csv",
                  delimiter=",")
air
```

NOx-Humidity

```python
#regression line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color = "#833131",
            alpha = 0.2,
            label='Data Points')

#line plot
plt.plot(x, line,
         color = "#128262",
         label=f'Regression Line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```
