# Measures of Central Tendency

```python
#Jaguar vs Panther
resistance =  np.array([
    [10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
    [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]
])

resistance
```

```python
np.shape(resistance)
```

```python
#Mean
res_mean = np.nanmean(resistance,axis=1)
res_mean[0]
```

```python
print("Jaguar Mean: " + str((res_mean[0]).round(2)) + " Ω")
```

```python
print(f'Jaguar: {res_mean[0]:.2f} Ω')
print(f'Panther: {res_mean[1]:.2f} Ω')
```

```python
#Median
res_median = np.nanmedian(resistance,axis=1)
res_median[0]
```

```python
print(f'Jaguar: {res_median[0]:.2f} Ω')
print(f'Panther: {res_median[1]:.2f} Ω')
```

```python
res_mode = stats.mode(resistance, axis=1)
res_mode
```

```python
jag_res = resistance[0]

print("Jaguar Mode: ")
for i in multimode(jag_res):
    print(f'{i} Ω')
```

```python
pan_res = resistance[1]

print("Panther Mode: ")
for i in multimode(pan_res):
    print(f'{i} Ω')
```

```python
#Jaguar vs Panther
resistance =  np.array([
    [10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
    [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]
])

resistance
```

```python
#Range
res_max = np.nanmax(resistance, axis=1)
res_min = np.nanmin(resistance, axis=1)

res_range = res_max - res_min
res_range
```

```python
print(f'Jaguar: {res_range[0]:.2f} Ω')
print(f'Panther: {res_range[1]:.2f} Ω')
```

```python
res_Q1 = np.nanpercentile(resistance,25,axis=1)
res_Q1
```

```python
res_Q3 = np.nanpercentile(resistance,75,axis=1)
res_Q3
```

```python
print(f'Jaguar IQR:  {res_Q3[0] - res_Q1[0]: .2f} Ω')
print(f'Panther IQR:  {res_Q3[1] - res_Q1[1]: .2f} Ω')
```

```python
res_var = np.nanvar(resistance, ddof =1,axis=1)

print(f'Jaguar Variance:  {res_var[0]:.2f} Ω')
print(f'Panther Variance:  {res_var[1]:.2f} Ω')
```

```python
res_std = np.nanstd(resistance, ddof =1,axis=1)

print(f'Jaguar Variance:  {res_std[0]:.2f} Ω')
print(f'Panther Variance:  {res_std[1]:.2f} Ω')
```
