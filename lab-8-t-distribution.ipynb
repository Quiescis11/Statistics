```python
#import and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Load Dataset
df = pd.read_csv(r'C:\Users\LENOVO\OneDrive\Documents\Github\Statistics\datasets\transformer-voltage-10-sample.csv',
                delimiter=",",
                index_col='Sample')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1-alpha/2,n-1)
t_critical
```

```python
#confidence interval
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)

print(f"CI = [{ci_lower:.2f}, {ci_upper:.2f}]") 
```

```python
#voltage distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

#x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

#normal distribution plot
x = np.linspace(x_min,x_max,100)
p = stats.norm.pdf(x,mean,std)
plt.plot(x,p,
          linewidth=2,
          color='brown')

#histogram
plt.hist(df['Voltage'], bins=5,
         density=True,
         alpha = 0.8,
         color='green')

plt.title('Voltage Distribution')
plt.ylabel('Probability Density')
plt.xlabel('Voltage (V)')

#confidence interval
plt.axvline(x=ci_lower,
            color ='black',
            linestyle = '--')

plt.axvline(x=ci_upper,
            color='black',
            linestyle='--')
plt.show()
```

# Shoe Inventory   

```python
#Load Dataset
df = pd.read_csv(r'datasets\shoe-inventory.csv',
                delimiter=",")
df
```

```python
df.shape
```

```python
# mean of size 6 
mean_size_6 = df.iloc[0,1:13].mean()
mean_size_6
```

```python
# standard deviation of size 6
std_size_6 = df.iloc[0,1:13].std()
std_size_6
```

```python
# size of 6 95% CI
n = df.shape[1]-1  # returns 13 in (17,13)

lower_6, upper_6 = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean_size_6,
    scale = std_size_6/np.sqrt(n)
)

print(f"CI = [{lower_6:.2f}, {upper_6:.2f}]") 
```

```python
# 25% Confidence Level
mean = df.iloc[0:17, 1:13].mean(axis=1)
std = df.iloc[0:17, 1:13].std(axis=1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)

print(f"CI = [{ci_lower:.2f}, {ci_upper:.2f}]") 
```

```python
df
```

```python
# fill NAN with 0
df['ci_lower'] = df['ci_lower'].fillna(0)
df['ci_upper'] = df['ci_upper'].fillna(0)
```

```python
# convert to int and round to 0

df['ci_lower'] = df['ci_lower'].astype(int).round()
df['ci_upper'] = df['ci_upper'].astype(int).round()
```
