```python
#import and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Load Dataset
df = pd.read_csv(r'C:\Users\LENOVO\OneDrive\Documents\Github\Statistics\transformer-voltage.csv',
                delimiter=",",
                index_col='Sample')
df
```

```python
alpha = 0.05
z_critical = stats.norm.ppf(1-alpha/2)
z_critical
```

```python
#confidence interval
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
    confidence=0.95,
    loc = df['Voltage'].mean(),
    scale = pop_std/np.sqrt(df['Voltage'].count())
)

print(f"CI = [{ci_lower:.2f}, {ci_upper:.2f}]") 
```

```python
#voltage distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

#x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

#normal distribution plot
x = np.linspace(x_min,x_max,100)
p = stats.norm.pdf(x,mean,std)
plt.plot(x,p,
          linewidth=2,
          color='brown')

#histogram
plt.hist(df['Voltage'], bins=5,
         density=True,
         alpha = 0.8,
         color='green')

plt.title('Voltage Distribution')
plt.ylabel('Probability Density')
plt.xlabel('Voltage (V)')

#confidence interval
plt.axvline(x=ci_lower,
            color ='black',
            linestyle = '--')

plt.axvline(x=ci_upper,
            color='black',
            linestyle='--')
plt.show()
```
